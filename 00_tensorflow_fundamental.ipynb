What we've created so far:

### Creating tensors with tf.Variable

```python
# Create the same tensor with tf.Variable() as above
changeable_tensor = tf.Variable([10, 7])
unchangeable_tensor = tf.constant([10, 7])
changeable_tensor, unchangeable_tensor
```

```python
# How about we try .assign()
changeable_tensor[0].assign(7)
changeable_tensor
```

```python
# Let's try change our unchangeable_tensor
unchangeable_tensor[0].assign(7)
```

Note: Rarely in practice will you need to decide whether to use tf.constant or tf.Variable to create tensors, as TensorFlow does this for you. However, if in doubt, use tf.constant and change it later if needed.

### Creating random tensor

Random tensors are tensors of some arbitrary size which contain radom numbers.

```python
# Create two random of a same tensor
random_1 = tf.random.Generator.from_seed(7) #set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(7)
random_2 = random_2.normal(shape=(3,2))

# Are they equal ?
random_1, random_2, random_1==random_2
```

### Getting information from tensors
- Shape: tensor.shape
- Rank: tensor.ndim
- Axis or dimension: tensor[0]
- Size

```python
# Create a rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros(shape = [2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attributes of our tensor:
print("Datatype of every element: ", rank_4_tensor.dtype)
print("Number of dimensions (rank): ", rank_4_tensor.ndim)
print("Shape of tensor: ", rank_4_tensor.shape)
print("Elements along the 0 axis: ", rank_4_tensor.shape[0])
print("Elements along the 1 axis: ", rank_4_tensor.shape[-1])
print("Total number of elements in our tensor: ", tf.size(rank_4_tensor))
print("Total number of elements in our tensor: ", tf.size(rank_4_tensor).numpy())
```

### Indexing tensors
Tensors can be indexed just like Python lists

```python
# Get the first 2 elements of each dimension
rank_4_tensor[:2, :2, :2, :2]
```

```python
# Get each dimension from each index except for the final one
rank_4_tensor[:-1, :-1, :-1, :-1]
```

```python
rank_4_tensor[:1, :, :1, :1]
```

```python
# Create a rank 2 tensor (2 dimension)
rank_2_tensor = tf.constant([[10, 7],
                             [3, 4]])
rank_2_tensor.shape, rank_2_tensor.ndim
```

```python
rank_2_tensor
```

```python
# Get the last item of each row of our rank 2 tensor
rank_2_tensor[:, -1]
```

```python
# Add in extra dimentsion to our rank 2 tensor
rank_3_tensor = rank_2_tensor[..., tf.newaxis]
rank_3_tensor
```

```python
# Alternative to tf.newaxis
tf.expand_dims(rank_2_tensor, axis = -1) #-1 means expand the final axis
```

```python
# Expand the 0 axis
tf.expand_dims(rank_2_tensor, axis = 0) #expand the 0-axis
```

**Matrix multiplication**

```python
# Matrix multiplication in tensor flow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
# Matrix multiplication with Python operator "@"
tensor @ tensor
```

```python
tensor.shape
```

```python
# Create a tensor (3,2)
X = tf.constant([[1, 2],
                [3, 4],
                [5, 6]])
Y = tf.constant([[7, 8],
                 [9, 10],
                 [11, 12]])
X, Y
```

```python
# Let's change the shape of Y
tf.reshape(Y, shape = (2,3))
```

```python
# Try to multiply X by reshaped Y
X @ tf.reshape(Y, shape = (2,3))
```

```python
tf.matmul(X, tf.reshape(Y, shape = (2,3)))
```

```python
tf.matmul(tf.reshape(X, shape=(2,3)), Y)
```

```python
# Can do the same with transpose
X, tf.transpose(X), tf.reshape(X, shape = (2,3))
```

```python
# Try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(X), Y)
```

```python
X, Y
```

**The dot product**
Matrix multiplication is also referred as dot product
You can perform matrix multiplication using:
- tf.matmul()
- tf.tensordot()

```python
# Perform the dot product on X and Y
tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
# Perfrom matrix multiplication between X and transpose Y
tf.matmul(X, tf.transpose(Y))
```

```python
# Perform matrix multiplication between X and reshape Y
tf.matmul(X, tf.reshape(Y, shape = (2,3)))
```

```python
# Check the values of Y, reshape Y and transposed Y
print("Normal Y: ")
print(Y, "\n")

print("Y reshaped to (2, 3): ")
print(tf.reshape(Y, shape=(2,3)), "\n")

print("Y transposed: ")
print(tf.transpose(Y))
```

### Changing the datatype of a tensor

```python
# Create a new tensor with default datatype (float32)
B = tf.constant([1.7, 7.4])
B.dtype
```

```python
C = tf.constant([7, 10])
C.dtype
```

```python
# Change from float32 to float16 (reduce precision)
D = tf.cast(B, dtype = tf.float16)
D, D.dtype
```

```python
# Change from int32 to float 32
E = tf.cast(C, dtype = tf.float32)
E, E.dtype
```

```python
E_float16 = tf.cast(E, dtype=tf.float16)
E_float16
```

### Aggregating tensors
Aggregating tensors = condensing them from multiple values down to a smaller amount of values

```python
# Get the absolute values
D = tf.constant([-7, -10])
D
```

```python
tf.abs(D)
```

Let's go through the following form of aggregations:
- Get the minimum
- Get the maximum
- Get the mean of a tensor
- Get the sum of a tensor

```python
import numpy as np
#Creating a random tensor with values betweens 0 and 100 of size 50
E = tf.constant(np.random.randint(0, 100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# Find the variance
E = tf.cast(E, dtype=tf.float32)
tf.math.reduce_variance(E)
```

```python
# Find the variance using tensorflow_probability
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# Find the standard deviation
tf.math.reduce_std(E)
```

### Find the positional maximum and minimum

```python
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Positional maximum
tf.math.argmax(F)
```

```python
# Index on out largest value position
F[tf.math.argmax(F)]
```

```python
# Find the max value of F
tf.reduce_max(F)
```

```python
# Check for quality
F[tf.math.argmax(F)] == tf.reduce_max(F)
```

```python
# Positional minimum
tf.math.argmin(F)
```

```python
# Find the minimum using index
F[tf.math.argmin(F)]
```
